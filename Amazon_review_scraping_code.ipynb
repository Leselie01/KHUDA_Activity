```python
from selenium import webdriver as wd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import requests
import re
import time
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup as bs
import pandas as pd
import os
from datetime import datetime

def dong_amazon_reviews(asin):
    
    options=Options()
    options.add_argument("--headless")
    options.add_argument("--window-size=1920x1080")
    options.add_argument("disable-gpu")
    # 여기 Path 부분에 Chromedriver.exe가 설치된 경로를 지정
    driver=wd.Chrome(executable_path = 'c:/informs/chromedriver.exe',options=options)
    
    url=f'https://www.amazon.com/Peter-Thomas-Roth-Temporary-Tightener/product-reviews/B0040QG46E/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
    #Driver Get Url
    driver.get(url)
    driver.implicitly_wait(10)
    res=driver.page_source
    obj=bs(res,'html.parser', from_encoding = 'utf-8')

    # 총 리뷰수 가져오기
    rev=obj.select('div.a-row.a-spacing-base.a-size-base')[0].get_text().strip().replace(',', '')
    rev=int(re.findall('\d+',rev)[1])

    titles=[]
    stars=[]
    contents=[]

    while len(titles) < rev:
        print(str(rev) + "개 중" + str(len(titles)) + "번째 데이터 크롤링 중..")
        try:
            time.sleep(1)
            source=driver.page_source
            bs_obj=bs(source,"html.parser", from_encoding = 'utf-8')

            #리뷰 타이틀 가져오기
            for i in bs_obj.findAll('a',{'data-hook':'review-title'}):
                titles.append(i.get_text().strip())

            #리뷰 내용 가져오기
            for a in bs_obj.findAll('span',{'data-hook':'review-body'}):
                contents.append(a.get_text().strip())

            #리뷰 평점 가져오기
            for u in bs_obj.findAll('i',{'data-hook':'review-star-rating'}):
                stars.append(int(u.get_text()[7]))

            for u in bs_obj.findAll('i',{'data-hook':'cmps-review-star-rating'}):
                stars.append(int(u.get_text()[7])) 

            if len(titles) == rev:
                break
        
            #다음 리뷰 페이지롤 넘어갈 수 있는 Next Button 클릭
            driver.find_element_by_xpath('//*[@id="cm_cr-pagination_bar"]/ul/li[2]').click()
            WebDriverWait(driver, 15).until(
            # 지정한 한개 요소가 올라면 웨이트 종료
                EC.presence_of_element_located((By.XPATH, '//*[@id="hmenu-container"]/span/input'))
            )
        except:
            continue
        
    driver.close()
    driver.quit()        
    df = pd.DataFrame({'Rating':stars,"Title":titles,"Body":contents})
    
    return df 

data = dong_amazon_reviews('B0040QG46E')

data.to_csv('AmazonReview_3.csv', encoding = "utf-8-sig")
```
